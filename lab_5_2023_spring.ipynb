```python
# Task 1
import math

class Circle:
    
    def __init__(self, radius):
        self.__radius = radius

    def setRadius(self, radius):
        self.__radius = radius
        
    def getRadius(self):
        return self.__radius
    
    def area(self):        
        return math.pi * ( self.__radius * self.__radius )
    


c1 = Circle(4)
print("First circle radius:" , c1.getRadius())
print("First circle area:" ,c1.area())

c2 = Circle(5)
print("Second circle radius:" ,c2.getRadius())
print("Second circle area:" ,c2.area())
```

```python
# Task 2 

class Triangle:
    
    def __init__(self, base, height):
        self.__base =base
        self.__height =height

    def setBase(self, base):
        self.__base = base
        
    def getBase(self):
        return self.__base

    def setHeight(self, height):
        self.__height = height
        
    def getHeight(self):
        return self.__height

    def area(self):
        a = (self.__base * self.__height)/2
        return a

  


t1 = Triangle(10, 5)
print("First Triangle Base:" , t1.getBase())
print("First Triangle Height:" , t1.getHeight())
print("First Triangle area:" ,t1.area())

t2 = Triangle(5, 3)
print("Second Triangle Base:" , t2.getBase())
print("Second Triangle Height:" , t2.getHeight())
print("Second Triangle area:" ,t2.area())
```

```python
class Train:
    def __init__(self, name, start, *stops, destination):
        self.name = name
        self.stops = [start, *stops, destination]
        self.passengers = []

    def addPassenger(self, *passengers):
        for p in passengers:
            self.passengers.append(p)
            p.currentStop = self.stops.index(p.start)
            p.fare = 0
            for i in range(p.currentStop, self.stops.index(p.destination)):
                p.fare += 100

    def allPassengerDetails(self):
        print(f"Welcome aboard on {self.name}")
        print(f"Start: {self.stops[0]}")
        print(f"Destination: {self.stops[-1]}")
        for p in self.passengers:
            print(f"Name: {p.name},Start: {p.start},Destination: {p.destination},Fair: ${p.fare}")

class Passenger:
    def __init__(self, name, start=None, destination=None):
        self.name = name
        self.start = start
        self.destination = destination
        self.currentStop = None
        self.fare = 0

    def __str__(self):
        return f"{self.name} welcome aboard"




t1 = Train('T1-Express','New York','Manhattan','Brooklyn',destination='Boston')
print("1========================")
p1 =Passenger("Naruto")
t1.addPassenger(p1)
p2 = Passenger("Sasuke","Manhattan","Boston")
p3 = Passenger("Hinata","Manhattan","Brooklyn")
print("2========================")
t1.addPassenger(p2,p3)
print("3========================")
t1.allPassengerDetails()
print("4========================")
t2 = Train('Europe-Express','London','Paris','Brussels',destination='Turkey')
print("5========================")
p4 =Passenger("Max","London","Brussels")
p5 = Passenger("Eleven","Paris","Turkey")
p6 = Passenger("Mike","Brussels","Turkey")
t2.addPassenger(p4,p5,p6)
print("6========================")
t2.allPassengerDetails()
```

```python
# Task 7

class BracuStudent:
    def __init__(self, name, home):
        self.name = name
        self.home = home
        self.bus_pass = False

    def get_pass(self):
        self.bus_pass = True
        print(f"{self.name} got the bus pass!")

    def show_details(self):
        pass_status = "True" if self.bus_pass else "False"
        print(f"Student Name: {self.name}\nLives in {self.home}\nHave Bus Pass? {pass_status}")

class BracuBus:
    def __init__(self, route, capacity=2):
        self.route = route
        self.capacity = capacity
        self.passengers = []

    def board(self, *students):
        for student in students:
            if student.bus_pass and student.home == self.route and len(self.passengers) < self.capacity:
                self.passengers.append(student.name)
                print(f"{student.name} boarded the bus.")
            elif not student.bus_pass:
                print("You don't have bus pass!")
            elif student.home != self.route:
                print("You got on wrong bus!")
            else:
                print("Bus is full!")

    def show_details(self):
        print(f"Bus Route: {self.route}\nPassengers Count: {len(self.passengers)} (Max: {self.capacity})")
        print(f"Passengers On Board: {self.passengers}")


st1 = BracuStudent("Afif", "Mirpur")
print("1===========================")
st2 = BracuStudent("Shanto", "Motijheel")
st3 = BracuStudent("Taskin", "Mirpur")
st1.show_details()
st2.show_details()
print("2===========================")
st3.show_details()
print("3===========================")
bus1 = BracuBus("Mirpur")
bus2 = BracuBus("Azimpur", 5)
bus1.show_details()
bus2.show_details()
print("4===========================")
st2.get_pass()
st3.get_pass()
print("5===========================")
st2.show_details()
st3.show_details()
print("6===========================")
bus1.board()
print("7===========================")
bus1.board(st1, st2)
print("8===========================")
st1.get_pass()
st2.home = "Mirpur"
st1.show_details()
st2.show_details()
print("9===========================")
bus1.board(st1, st2, st3)
print("10===========================")
bus1.show_details()
```

```python
class Library:
    def __init__(self, name, books):
        self.name = name
        self.borrowers = {}
        self.books = books

    def details(self):
        print(f"{self.name} Library details")
        print("Borrower details:")
        print(self.borrowers)
        print("Books availability:")
        print(self.books)

class Reader:
    def __init__(self, name):
        self.name = name
        self.borrowed_books = {}

    def borrow(self, library, *book_types):
        for book_type in book_types:
            if book_type in library.books and library.books[book_type] > 0:
                if self.name in library.borrowers:
                    if len(self.borrowed_books) < 5:
                        self.borrowed_books[book_type] = self.borrowed_books.get(book_type, 0) + 1
                        library.books[book_type] -= 1
                        library.borrowers[self.name] = sum(self.borrowed_books.values())
                        print(f"{book_type} book is borrowed successfully.")
                    else:
                        print("You cannot borrow more than 5 books.")
                        break
                else:
                    self.borrowed_books[book_type] = 1
                    library.books[book_type] -= 1
                    library.borrowers[self.name] = sum(self.borrowed_books.values())
                    print(f"{book_type} book is borrowed successfully.")
            else:
                print(f"{book_type} books are not available at the moment.")

    def readerInfo(self, book_type=None):
        if book_type:
            if book_type in self.borrowed_books:
                print(f"{self.name}, you have {self.borrowed_books[book_type]} {book_type} book(s) with you.")
            else:
                print(f"{self.name}, you have no {book_type} book(s) with you.")
        else:
            print(f"{self.name}, you have {sum(self.borrowed_books.values())} book(s) with you.")





L1=Library('Dhaka',{'Arts':15,'Fiction':135,'Politics':2,'Science':11,'Poetry':15})
L1.details()
print("1----------------------")
r1=Reader('Aladdin')
r1.borrow(L1,'Arts','Fiction','Fiction','Politics')
print("2----------------------")
r1.borrow(L1,'Politics','Fiction')
print("3----------------------")
r1.readerInfo()
print("4----------------------")
r1.readerInfo('Fiction')
print("5----------------------")
L1.details()
print("6----------------------")
r2=Reader('Jasmine')
r2.borrow(L1,'Politics','Poetry')
print("7----------------------")
r2.readerInfo()
print("8----------------------")
L1.details()

```
