```python
#TASK_1

class Student:
  ID = 0
  def __init__(self, name, dept, age, cg):
    self.name = name
    self.dept = dept
    self.age = age
    self.cg = cg
    Student.ID+=1
  def get_details(self):
    print('ID:', Student.ID)
    print('Name:', self.name)
    print('Department:', self.dept)
    print('Age:', self.age)
    print('CGPA:', self.cg)
  @classmethod
  def from_String(cls, string):
    cls.string = string
    name, dept, age, cg = cls.string.split('-')
    newstudent = cls(name, dept, age, cg)
    return newstudent

s1 = Student("Samin", "CSE", 21, 3.91)
s1.get_details()
print("-----------------------")
s2 = Student("Fahim", "ECE", 21, 3.85)
s2.get_details()
print("-----------------------")
s3 = Student("Tahura", "EEE", 22, 3.01)
s3.get_details()
print("-----------------------")
s4 = Student.from_String("Sumaiya-BBA-23-3.96")
s4.get_details()
print("-----------------------")
print('Class variables are declared inside a class but outside any method. So,\nit can be accessed by any member of the class. Whereas, Instance variables\nare declared within a constructor or a method inside a class, which can be\naccessed within that constructor or that method.')
print("-----------------------")
print('Class methods are bound to a class, and it can access the instance\nvariables within the whole class. But instance methods are\nblueprints of objects that are bound to the objects.')
```

```python
#TASK_2

class Assassin:
  assassinCount = 0
  def __init__(self, name, rate):
    self.name = name
    self.rate = rate
    Assassin.assassinCount+=1
  def printDetails(self):
    print('Name:',self.name)
    print('Success rate:',str(self.rate)+'%')
    print('Total number of assassins:',Assassin.assassinCount)  
  @classmethod
  def failureRate(cls, frname, frate):
    psrate = (100-frate)
    obj1 = cls(frname, psrate)
    return obj1    
  @classmethod
  def failurePercentage(cls, fpname, fpate):
    prate = (100-fpate)
    obj2 = cls(fpname, prate)
    return obj2


john_wick = Assassin('John Wick', 100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate("Nagisa", 20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage('Akabane', 10)
akabane.printDetails()
```

```python
#TASK_7

class Cat:
  Number_of_cats = 0
  def __init__(self, color, action):
    self.color = color
    self.action = action
    Cat.Number_of_cats+=1
  @classmethod
  def no_parameter(cls):
    color1 = 'White'
    action1 = 'sitting'
    object1 = cls(color1, action1)
    return object1
  @classmethod
  def first_parameter(cls, color2):
    action2 = 'sitting'
    object2 = cls(color2, action2)
    return object2
  @classmethod
  def second_parameter(cls, action3):
    color3 = 'Grey'
    object3 = cls(color3, action3)
    return object3
  def changeColor(self, color4):
    self.color = color4
  def printCat(self):
    print(f'{self.color} cat is {self.action}')       


print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
#TASK_8

import math
class Cylinder:
  height = 18
  radius = 5
  

  def __init__(self, radius, height):
    self.height = height
    self.radius = radius
    
    print(f'Default radius={Cylinder.radius} and height={Cylinder.height}')
    Cylinder.height = self.height
    Cylinder.radius = self.radius
    print(f'Upadated radius={Cylinder.radius} and height={Cylinder.height}')
  @staticmethod
  def area(radius, height):
    area = (2*math.pi*(radius)**2) + 2*math.pi*radius*height
    print(f'Area: {area}')
  @staticmethod
  def volume(radius, height):
    volume = math.pi*(radius**2)*height
    print(f'Volume: {volume}')
  @classmethod
  def swap(cls, radius, height):
    swapped_h = radius
    swapped_r = height
    object1 = cls(swapped_r, swapped_h)
    return object1  
  @classmethod
  def changeFormat(cls, string):
    lst1 = string.split('-')
    newrad = float(lst1[0])
    newhei = float(lst1[1])
    object2 = cls(newrad, newhei)
    return object2

c1 = Cylinder(0, 0)
Cylinder.area(c1.radius, c1.height)
Cylinder.volume(c1.radius, c1.height)
print("===============================")
c2 = Cylinder.swap(8, 3)
c2.area(c2.radius, c2.height)
c2.volume(c2.radius, c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius, c3.height)
c3.volume(c3.radius, c3.height)
print("===============================")
Cylinder(0.3, 5.56).area(Cylinder.radius, Cylinder.height)
print("===============================")
Cylinder(3, 5).volume(Cylinder.radius, Cylinder.height)
```
