```python
#TASK_2

limit = int(input('How many lists do you want: '))

combined_list = []

for i in range(limit):
    user_input = input('Enter a list: ')
    string_list = user_input.split(' ')
    #print(string_list)
    
    integer_list = []
    
    for j in string_list:
        integer_list.append(int(j))
    #print(integer_list)
        
    combined_list.append(integer_list)
    #print(combined_list)
    
respective_list = []
temporary_sum = 0
sum = 0
    
for elements in combined_list:
    for element in elements:
        temporary_sum += element
        
    if temporary_sum > sum:
        sum = temporary_sum
        temporary_sum = 0
        respective_list = elements
        
print(sum)
print(respective_list)
```

```python
#TASK_4

def ub_jumper(user_input):
    ub_checker = False
    ub_checker = user_input.split(" ")
    
    my_list =[]
    length = len(ub_checker)
    
    for i in range(length):
        number = int(ub_checker[i + 1])- int(ub_checker[i])
        
        if (number < 0):
            number = -1 * number
            my_list.append(number)
        else:
            my_list.append(number)
            
        if (i == (length - 2)):
            break
            
    my_list.sort()
    list_length =len(my_list)
    
    if (list_length == (length - 1)):
        count = 0
        
        for i in range(list_length):
            count += 1
            
            if (count == my_list[i]):
                ub_checker = True
            else:
                ub_checker = False
                break
                
    if (ub_checker == True):
        print('UB Jumper')
    else:
        print('Not UB Jumper')
        
for j in range(99999999999):
    user_input = input("Enter a number sequence: ")
    
    if (user_input == "STOP"):
        break
    else:
        ub_jumper(user_input)
```

```python
#TASK_5

string = input('Enter a string: ')

main_list = []

for element in string:
    main_list.append(element)

lowercase_list = []
uppercase_list = []
odd_list = []
even_list = []

for element in main_list:
    
    if (97 <= ord(element) <= 122):
        lowercase_list.append(element)
    elif (65 <= ord(element) <= 90):
        uppercase_list.append(element)
        
    elif (48 <= ord(element) <= 57):
        number = int(element)
        
        if (number % 2 != 0): 
            odd_list.append(number)
        elif (number % 2 == 0):
            even_list.append(number)

lowercase = ''
uppercase = ''
odd = ''
even = ''

for element in sorted(lowercase_list):
    lowercase += element
for element in sorted(uppercase_list):
    uppercase += element
for element in sorted(odd_list):
    odd += str(element)
for element in sorted(even_list):
    even += str(element)
    
print(lowercase + uppercase + odd + even)   
```
